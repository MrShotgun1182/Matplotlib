# Matplotlib

## import lib:

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import seaborn as sns
from pylab import *
```

## my plots:

### y1 = x^2 & y2 = x^3:

```python
x = np.linspace(-5, 5, 200)
y1 = x ** 2
y2 = x ** 3

fig, ax = plt.subplots()
ax.plot(x, y1, 'r')
ax.plot(x, y2, 'k--')
ax.set_xlabel(r'$x$', size= 18)
ax.set_ylabel(r'$y$', size= 18)
ax.legend(["y_1 = x^2", "y_2 = x^3"])
plt.show()
```

### y1 = sin(x) & y2 = cos(x):

```python
x  = np.arange(0, 5*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y1, "b-")
plt.plot(x, y2, color='red')
plt.xlabel("x-Axis", fontsize='large')
plt.ylabel("y-Axis", fontsize='large')
plt.legend(['y1 = sin(x)', 'y2 = cos(x)'], loc='upper center', fontsize='large')
```

```python
plt.figure(figsize=(12,6))
x = np.arange(0, 5*np.pi, 0.1)
y1 = sin(x)
y2 = cos(x)
plt.subplot(2,1,1)
plt.plot(x, y1, "b-")
plt.subplot(2,1,2)
plt.plot(x, y2, "r.")
plt.tight_layout() # for float
plt.show()
```
