# Matplotlib

## import lib:

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import seaborn as sns
from pylab import *
```

## my plots:

### y1 = x^2 & y2 = x^3:

```python
x = np.linspace(-5, 5, 200)
y1 = x ** 2
y2 = x ** 3

fig, ax = plt.subplots()
ax.plot(x, y1, 'r')
ax.plot(x, y2, 'k--')
ax.set_xlabel(r'$x$', size= 18)
ax.set_ylabel(r'$y$', size= 18)
ax.legend(["y_1 = x^2", "y_2 = x^3"])
plt.show()
```

### y1 = sin(x) & y2 = cos(x):

```python
x  = np.arange(0, 5*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y1, "b-")
plt.plot(x, y2, color='red')
plt.xlabel("x-Axis", fontsize='large')
plt.ylabel("y-Axis", fontsize='large')
plt.legend(['y1 = sin(x)', 'y2 = cos(x)'], loc='upper center', fontsize='large')
```

```python
plt.figure(figsize=(12,6))
x = np.arange(0, 5*np.pi, 0.1)
y1 = sin(x)
y2 = cos(x)
plt.subplot(2,1,1)
plt.plot(x, y1, "b-")
plt.subplot(2,1,2)
plt.plot(x, y2, "r.")
plt.tight_layout() # for float
plt.show()
```

### Account circulation plt:

```python
x = np.arange(1, 21)
y1 = np.random.uniform(0.1, 0.7, 20)
y2 = np.random.uniform(0.1, 0.7, 20)

plt.figure(figsize=(16,8))
plt.bar(x, +y1)
plt.bar(x, -y2, color='red')

for x, y, y2 in zip(x, y1, y2):
    plt.text(x, y+0.05, '%.2f' %y, fontsize=12, ha='center', va='bottom')
    plt.text(x, -y2-0.1, '%.2f' %y2, fontsize=12, ha='center', va='bottom')

plt.xlim(0,21)
plt.ylim(-1.25, +1.25)
plt.show()
```

### Bar chart:

```python
id1 = np.arange(1,10)
score = np.arange(20, 110, 10)

plt.figure(figsize=(8,5))
ax = plt.axes()
ax.set_facecolor("#ECF0F1")
plt.bar(id1, score, color='#FFA726')
plt.xlabel('Student ID')
plt.ylabel('Score')
for x,y in zip(id1, score):
    plt.text(x, y+0.05, '%.2f' %y, fontsize=10, ha='center', va='bottom')
plt.show()
```

### Pie Plot:

```python
plt.figure(figsize=(7,7))
area = [48,30,20,15]
labels = ['Low','Medium','High','Critical']
color = ['#88C34A','#D4E157','#FFB300','#FF7043']
plt.pie(area, labels=labels, colors=color, startangle=180)
plt.show()
```

```python
plt.figure(figsize=(7,7))
area = [48,30,20,15]
labels = ['Low','Medium','High','Critical']
color = ['#7CB342','#C0CA33','#FFB300','#F57C00']
plt.pie(area, labels=labels, colors=color, startangle=180, shadow=True, autopct='%.2f')
plt.show()
```

### Scater plt:

```python
plt.figure(figsize=(12,6))
x = np.random.normal(0, 1, 150)
y = np.random.normal(0, 1, 150)
plt.scatter(x[:75], y[:75], color='red')
plt.scatter(x[75:], y[75:])
plt.plot([-1, -1], [-3, 3], 'green')
plt.plot([1, 1], [-3, 3], 'green')
plt.show()
```

### stackplot:

```python
x = np.arange(1,6)
y = [[1,5,9,13,17], [2,8,11,15,19], [3,10,13,19,22], [3,12,15,21,25]]
color = ['#EFE9D5','#71BBB2','#497D74','#27445D']
plt.figure(figsize=(8,6))
plt.stackplot(x, y, labels=['y1','y2','y3','y4'], colors=color)
for x,y in zip(x, y[0]):
    plt.text(x, y, y, fontsize=8, ha='center', va='bottom')
plt.legend(loc='upper left')
plt.show()
```

### histogram plt:

```python
x = np.random.normal(size=200)

plt.hist(x, color='yellowgreen', bins=20)
plt.gca().set(title='Histogram', ylabel='Frequency')
plt.show()
```
