# Matplotlib

## import lib:

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import seaborn as sns
from pylab import *
```

## my plots:

### y1 = x^2 & y2 = x^3:

```python
x = np.linspace(-5, 5, 200)
y1 = x ** 2
y2 = x ** 3

fig, ax = plt.subplots()
ax.plot(x, y1, 'r')
ax.plot(x, y2, 'k--')
ax.set_xlabel(r'$x$', size= 18)
ax.set_ylabel(r'$y$', size= 18)
ax.legend(["y_1 = x^2", "y_2 = x^3"])
plt.show()
```
